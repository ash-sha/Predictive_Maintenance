```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import plotly.graph_objects as go
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("/Users/aswathshakthi/PycharmProjects/MLOps/Predictive_maintenance/data/processed/df_train.csv")
df_train
```

```python
df_test = pd.read_csv("/Users/aswathshakthi/PycharmProjects/MLOps/Predictive_maintenance/data/processed/df_test.csv")
df_test
```

```python
# Assume df_train and df_test are already loaded DataFrames with 'RUL' as the target variable
X_train = df_train.drop(columns=['RUL'])
y_train = df_train['RUL']
X_test = df_test.drop(columns=['RUL'])
y_test = df_test['RUL']
```

```python
# Normalisation

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform both training and test data
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)

```

### Model

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import RFE

# Polynomial Feature Transformation
degree = 2  # Degree of polynomial features
poly = PolynomialFeatures(degree,include_bias=False)
X_train_poly = poly.fit_transform(X_train_normalized)
X_test_poly = poly.transform(X_test_normalized)

# Train the Model
model = LinearRegression(fit_intercept=True,n_jobs=-1)
rfe = RFE(model, n_features_to_select=10)
fit = rfe.fit(X_train_poly, y_train)

# Display selected features
selected_features_rfe = poly.get_feature_names_out()[fit.support_]
print("Selected features with RFE:", selected_features_rfe)
```

```python
# Transform training and test data to keep only selected features
X_train_rfe = rfe.transform(X_train_poly)
X_test_rfe = rfe.transform(X_test_poly)
```

```python
# Train the model with selected features
model.fit(X_train_rfe, y_train)
```

```python
import joblib

# Assuming `model` is your trained model
joblib.dump(model, '/Users/aswathshakthi/PycharmProjects/MLOps/Predictive_maintenance/models/polyreg.pkl')
```

```python
# Load the model
model = joblib.load('/Users/aswathshakthi/PycharmProjects/MLOps/Predictive_maintenance/models/polyreg.pkl')
```

```python

# Make predictions
y_train_pred = model.predict(X_train_rfe)
y_test_pred = model.predict(X_test_rfe)
```

### Evaluation

```python
#MSE -train
def mse_tr(y_train, y_train_pred):
    mse = mean_squared_error(y_train, y_train_pred)
    return (f'Mean Squared Error - Train: {mse}')
mse_tr(y_train, y_train_pred)
```

```python
# R-squared evaluation - train
def r2_tr(y_train, y_train_pred):
    r_squared = r2_score(y_train, y_train_pred)
    return (f'R-squared - Train: {r_squared}')
r2_tr(y_train, y_train_pred)
```

```python
#MSE -test
def mse_tt(y_test, y_test_pred):
    mse = mean_squared_error(y_test, y_test_pred)
    return (f'Mean Squared Error - Test: {mse}')
mse_tt(y_test, y_test_pred)
```

```python
# R-squared evaluation - test
def r2_tt(y_test, y_test_pred):
    r_squared = r2_score(y_test, y_test_pred)
    return (f'R-squared - Test: {r_squared}')
r2_tt(y_test, y_test_pred)
```
